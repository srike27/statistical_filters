<h1> Moving Average Filter </h1>
<p>The moving average filter averages out the past $ n $ values and produces an output as that. It can be used to remove large discontinuities in the values. It is an implicit low pass filter</p>

```python
# header initializations #
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def movavg(inp,n):
    N = inp.size
    op = np.zeros(inp.size)
    if N < n:
        return np.mean(inp)
    else:
        for i in range(op.size-n+1):
            op[int(i+n/2)] = np.mean(inp[i:i+n-1:1])
        return op
```

```python
a = np.sin(np.array(range(70))*np.pi/100.0) # signal
c = 0.1*np.sin(np.array(range(70))*40*np.pi/100)# noise
t = a+c # adding signal and noise
b = movavg(t,10)
ax = plt.plot(a,label = 'true signal', color = 'black')
ax = plt.plot(t,label = 'noisy signal', color = 'red')
ay = plt.plot(b,label= 'filterd output', color = 'cyan')
plt.legend(loc = 'upper left')
plt.show()
```
