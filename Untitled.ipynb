<h1> Weighted moving average </h1>
<p> The weighted moving average allows the user to give more importance to recent values and lesser importance to older values this can be of good use in some applications. It basically takes a weighted mean ove the past $ n $ values of the signal </p>

```python
# header initializations #
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def wmovavg(inp,n,alpha):
    N = inp.size
    op = np.zeros(inp.size)
    s = (1-alpha**n)/(1-alpha)
    print s
    if N < n:
        return np.mean(inp)
    else:
        sum = 0
        for i in range(op.size-n):)
            sum = 0
            for j in range(n):
                op[int(i+n/2)] = 1.0*op[int(i+n/2)] + 1.0*(alpha**j)*inp[i+j]
            op[int(i+n/2)] = op[int(i+n/2)]/s
        return op
```

```python
a = np.sin(np.array(range(70))*np.pi/100.0) # signal
c = 0.1*np.sin(np.array(range(70))*40*np.pi/100)# noise
t = a+c # adding signal and noise
b = wmovavg(t,10,0.9)
ax = plt.plot(a,label = 'true signal', color = 'black')
ax = plt.plot(t,label = 'noisy signal', color = 'red')
ay = plt.plot(b,label= 'filterd output', color = 'cyan')
plt.legend(loc = 'upper left')
plt.show()
```
